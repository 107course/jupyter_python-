face detection：人脸检测：检测视频或者图像中有没有人脸，人脸在哪里  
face alignment: 人脸配准  
face verification:    人脸验证，1：1的比对，你说你是A，那就拿你的人脸和A的人脸进行比对，给出YES or NO  
face identification：人脸识别，1：N的比对，不知道你是谁，直接和数据库里的人脸比对，得出一个结果  
face recognize：也称为人脸识别，更为宽泛一些的概念，可以包括验证、识别、检测、配置等  

face labeling：人脸标注  
face editing：人脸编辑，对人脸的一些处理，比如人脸美化应该就属于编辑  
face hallucination：人脸超分辨率
